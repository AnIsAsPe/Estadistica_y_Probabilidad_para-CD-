```
alarm = gum.loadBN("alarm.bif")
alarm
```

```
#¿cuántas variables y dependencias entre variables tenemos?
print(alarm)
```

```

nombre = [alarm.variable(i).name() for i in range(37)]
categorias = [alarm.variable(i).labels() for i in range(37)]

variables = pd.DataFrame({'nombre':nombre,
                          'categorias':categorias})
variables['categorias'] = variables['categorias'].astype(str)
variables['categorias'] = variables['categorias'].str.replace('[(\')]', ' ')
variables.head()
```

```
from bs4 import BeautifulSoup  
import requests
```

```
# url del que bajaremos la información
url = 'https://www.bnlearn.com/documentation/man/alarm.html'

#usamos requests para guardar la información de la url
respuesta = requests.get(url)

soup = BeautifulSoup(respuesta.text, 'html.parser')
```

```
variables = soup.find('div',{ "id" :"inner2"}, class_="blocks").find('ul')

nombres=[]
descripcion=[]
for var in variables.find_all('li'):
  nombres.append(var.find('code').text)
  descripcion.append(var.find('em').text)

des = pd.DataFrame({'Var':nombres,
                    'Descripcion':descripcion})
des.index= des.index+1
des
```

```
diccionario_alarm = pd.merge(nombres, variables, how='outer', on="nombre")
print(diccionario_alarm.shape)
diccionario_alarm
```

```
diccionario_alarm = diccionario_alarm.rename( columns={'descripcion_in':'descripcion'})
diccionario_alarm.head()
```
