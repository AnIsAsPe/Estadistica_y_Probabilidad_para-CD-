```python
import numpy as np
from numpy import exp
from scipy.special import factorial

import matplotlib.pyplot as plt
```

# Simulaci칩n binomial

## Bernoulli n = 1

```python
# Bernoulli n = 1
n, p = 1, 0.3
size = 1000
s = np.random.binomial(n, p, size)  #muestra de tama침o size con la preferencia de el candidato con .3
plt.hist( s,   weights = np.ones_like(s)/size)
plt.xticks([0,1])
plt.show()
print(s)
```

### Funci칩n de probabilidad

$$P(x)=p^{x}(1-p)^{1-x}$$  

```python
bernoulli_pmf = lambda x,  p:  p**x * (1-p)**(1-x)
xs = np.array([0,1])
p = 0.3
fig, ax = plt.subplots(figsize=(8, 6))


ax.plot(xs, bernoulli_pmf(xs,  p),
            label='$p={:.2f}$'.format( p),
            marker='o', alpha=0.8,
            linestyle = 'none',
            markersize = 14)

ax.bar(xs, bernoulli_pmf(xs,  p),
            width=.1,
            alpha=0.6)

ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$P$', fontsize=14)

plt.xticks([0,1])
ax.legend(fontsize=14)

plt.show()
```

### Funci칩n de probabilidad

$$P(x)= \frac{{n!}}{(n-x)!x!} p^{x}(1-p)^{n-x}$$  

```python
binom_pmf = lambda x, n, p: factorial(n) / (factorial(n-x)*factorial(x)) * p**x * (1-p)**(n-x)
n = 10
xs = range(0, n+1)

fig, ax = plt.subplots(figsize=(8, 6))

for p in [0.1, 0.5, 0.9]:
    distribution = []
    for x_i in xs:
        distribution.append(binom_pmf(x_i, n, p))
    ax.plot(xs, distribution,
            label='$n={}$, $p={:.2f}$'.format(n, p),
            marker='o', alpha=0.8,
            linestyle = '--',
            markersize = 14)

    ax.bar(xs,distribution,
            width=.4,
            alpha=0.3)

ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$P$', fontsize=14)
ax.axis(xmin=0, ymin=0)
plt.xticks(range(0,n+1))
ax.legend(fontsize=14)

plt.show()
```

## Poisson cuando ${n \to \infty}$
### ${p \to 0}$

La distribuci칩n Poisson es la distribuci칩n de bernulli, para eventos raros, cuando hacemos suficientes experimientos

```python
n, p = 100000, 0.00002
size = 1000000
s = np.random.binomial(n, p, size)  # muestra del tama침o size con la cantidad de  personas por cada  10000 que tengan una enfermedad rara (p=0.0001)
plt.hist( s, bins=30,  weights = np.ones_like(s)/size)
plt.show()
print(s[0:100])
```
