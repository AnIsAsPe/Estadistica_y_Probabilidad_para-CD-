```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
medias = [165, 68]
covarianza = [[11, 14], [14, 36]]
n = 10000

altura, peso = np.random.multivariate_normal(mean = medias,
                                             cov = covarianza,
                                             size = n).T

plot = sns.jointplot(x=altura, y=peso,
                     kind = 'scatter',
                     color = 'darkcyan',
                     s = 2)
plot.ax_marg_x.set_xlim(155, 175)
plot.ax_marg_y.set_ylim(40, 100)
plt.show()
```

# Variables en Regresión

```python
# Variables en regresión

x2  = np.random.normal(loc=0, scale=1, size = n)
y2 = 0.4 + 1.5 * x2 + np.random.normal(loc= 0, scale = 0.3, size=n)


plot = sns.jointplot(x=x2, y= y2, kind = 'scatter', color = 'darkcyan', s = 2)
plot.ax_marg_x.set_xlim(-5, 5)
plot.ax_marg_y.set_ylim(-5, 5)

M_covarianza2 = np.cov(x2, y2)
print(M_covarianza2.round(2))
plt.show()
```

```python
medias = [3, -3]
covarianza =[[10, -5], [-5, 4]]
rv1, rv2 = np.random.multivariate_normal(mean = medias, cov = covarianza, size = n).T
plot = sns.jointplot(x=rv1, y= rv2, kind = 'scatter', color = 'darkcyan', s = 2 )
plot.ax_marg_x.set_xlim(-10, 15)
plot.ax_marg_y.set_ylim(-10, 6)
plt.show()
```

# Generación de vectores en más de dos dimensiones

```python
medias = [-3, 6, 0]
covarianza =[[10, 4, -2], [4, 5, -1], [-2, -1, 7]]
rv1, rv2, rv3 = np.random.multivariate_normal(mean = medias, cov = covarianza, size = n).T
datos = pd.DataFrame({'Variable1': rv1, 'Variable2':rv2, 'Variable3':rv3})
datos
```
