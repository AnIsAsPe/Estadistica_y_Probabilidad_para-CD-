```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
medias = [172, 68]
covarianza = [[11, 9], [9, 10]]
n = 10000

altura, peso = np.random.multivariate_normal(mean = medias,
                                             cov = covarianza,
                                             size = n).T

sns.jointplot(x=altura, y=peso, kind = 'scatter', color = 'darkcyan', s = 2)
plt.show()
```

# Variables en Regresión

```python
# Variables en regresión

x2  = np.random.normal(loc=5, scale=1.2, size = n)
y2 = 6 - 2 * x2 + np.random.normal(size=n)


sns.jointplot(x=x2, y= y2, kind = 'scatter', color = 'darkcyan', s = 2 )
M_covarianza2 = np.cov(x2, y2)
print(M_covarianza2.round(2))
plt.show()
```

```python
medias = [3, 16]
covarianza =[[10, 0], [0, 4]]
rv1, rv2 = np.random.multivariate_normal(mean = medias, cov = covarianza, size = n).T
sns.jointplot(x=rv1, y= rv2, kind = 'scatter', color = 'darkcyan', s = 2 )
```

# Generación de vectores en más de dos dimensiones

```python
medias = [-3, 6, 0]
covarianza =[[10, 4, -2], [4, 5, -1], [-2, -1, 7]]
rv1, rv2, rv3 = np.random.multivariate_normal(mean = medias, cov = covarianza, size = n).T
datos = pd.DataFrame({'Variable1': rv1, 'Variable2':rv2, 'Variable3':rv3})
datos
```
